```python
def tokenize(data,n):
    FGMatch = []
    text = "".join(data)
    
    templist = []
    tt = ngrams(data,n)
    for aa in tt:                  
        i = "".join(aa)
        temp = ""
        
        ss = MatchText(i)
        if ss['flag'] == True:
            #print(ss['val'])
            FGMatch.append(ss['val'])            
            temp = text.replace(ss['val'],'')
            text = temp
            templist.append(text)
            
        else:
            templist.append(text)
            continue
    
    try: 
        result = {'dd':segment(templist[-1]),'mm':FGMatch}
    except:
        result = {'dd':data,'mm':FGMatch}
    return result
```

```python
def segment(text):    
    text = re.sub(r'(?:(?<!္)([က-ဪဿ၊-၏]|[၀-၉]+|[^က-၏]+)(?![ှျ]?[့္်]))',r's\1',text).strip('s').split('s')   
    temp = text
    
    return temp
```

```python
def MatchText(text):
    
    wordVal = {'flag':False, 'val':''}
    for w in matchlist:
        if text == w:
            wordVal['flag'] = True
            wordVal['val'] = text
        
    return wordVal
```
